# Using AG2 with Elasticsearch

This notebook demonstrates how to use [AG2](https://ag2.ai) (formerly AutoGen) with Elasticsearch for multi-agent job-candidate matching using hybrid search (BM25 + ELSER semantic via RRF).

Five agents collaborate in a GroupChat to retrieve candidates, research job offers, match them, and verify the results.

## Creating mappings and inference endpoint

```python
try:
    _client.options(
        request_timeout=60, max_retries=3, retry_on_timeout=True
    ).inference.put(
        task_type="sparse_embedding",
        inference_id="jobs-candidates-inference",
        body={
            "service": "elasticsearch",
            "service_settings": {
                "adaptive_allocations": {"enabled": True},
                "num_threads": 1,
                "model_id": ".elser_model_2",
            },
        },
    )

    print("Inference endpoint created successfully.")

except Exception as e:
    print(
        f"Error creating inference endpoint: {e.info['error']['root_cause'][0]['reason']}"
    )
```

```python
try:
    _client.indices.create(
        index="available-candidates",
        body={
            "mappings": {
                "properties": {
                    "candidate_name": {
                        "type": "text",
                        "copy_to": "semantic_field",
                    },
                    "position_title": {
                        "type": "text",
                        "copy_to": "semantic_field",
                    },
                    "profile_description": {
                        "type": "text",
                        "copy_to": "semantic_field",
                    },
                    "expected_salary": {
                        "type": "text",
                        "copy_to": "semantic_field",
                    },
                    "skills": {
                        "type": "keyword",
                        "copy_to": "semantic_field",
                    },
                    "semantic_field": {
                        "type": "semantic_text",
                        "inference_id": "jobs-candidates-inference",
                    },
                }
            }
        },
    )

    print("Index created successfully.")

except Exception as e:
    print(
        f"Error creating index: {e.info['error']['root_cause'][0]['reason']}"
    )
```

## AG2 (formerly AutoGen)

### LLM configuration

```python
llm_config = LLMConfig(
    {"model": "gpt-4o-mini", "api_key": os.environ["OPENAI_API_KEY"]}
)
```

### Tools

```python
def elasticsearch_hybrid_search(
    question: Annotated[str, "Search query for finding candidates"],
) -> str:
    """Search in Elasticsearch using hybrid search (BM25 + semantic via RRF)."""

    response = _client.search(
        index="available-candidates",
        body={
            "_source": {
                "includes": [
                    "candidate_name",
                    "position_title",
                    "profile_description",
                    "expected_salary",
                    "skills",
                ],
            },
            "size": 10,
            "retriever": {
                "rrf": {
                    "retrievers": [
                        {
                            "standard": {
                                "query": {"match": {"position_title": question}}
                            }
                        },
                        {
                            "standard": {
                                "query": {
                                    "semantic": {
                                        "field": "semantic_field",
                                        "query": question,
                                    }
                                }
                            }
                        },
                    ]
                }
            },
        },
    )

    hits = response["hits"]["hits"]
    if not hits:
        return "No candidates found."

    return json.dumps([hit["_source"] for hit in hits], indent=2)


def search_in_internet(
    query: Annotated[str, "Search query for finding job offers"],
) -> str:
    """Search the internet using Serper API for job offers."""

    url = "https://google.serper.dev/search"
    headers = {
        "X-API-KEY": os.environ["SERPER_API_KEY"],
        "Content-Type": "application/json",
    }

    payload = json.dumps({"q": query})
    response = requests.post(url, headers=headers, data=payload, timeout=30)
    response.raise_for_status()
    original_results = response.json()

    related_searches = original_results.get("relatedSearches", [])
    original_organics = original_results.get("organic", [])

    for search in related_searches:
        related_query = search.get("query", "")
        if not related_query:
            continue
        payload = json.dumps({"q": related_query})
        resp = requests.post(url, headers=headers, data=payload, timeout=30)
        if resp.ok:
            original_organics.extend(resp.json().get("organic", []))

    return json.dumps(original_organics[:20], indent=2)
```

### Agents

Five agents collaborate via GroupChat with LLM-driven speaker selection:
- **Admin** — initiates the process and terminates on completion
- **Retriever** — searches Elasticsearch for candidate profiles
- **Researcher** — searches the internet for job offers
- **Matcher** — matches candidates with suitable job offers
- **Critic** — verifies match quality and signals completion

```python
admin = ConversableAgent(
    name="Admin",
    system_message=(
        "You are a human administrator. "
        "Your role is to interact with agents and tools to execute tasks efficiently. "
        "Execute tasks and agents in a logical order, ensuring that all agents perform "
        "their duties correctly. All tasks must be approved by you before proceeding."
    ),
    human_input_mode="NEVER",
    llm_config=llm_config,
    is_termination_msg=lambda msg: msg.get("content") is not None
    and "TERMINATE" in msg["content"],
)

retriever = ConversableAgent(
    name="Retriever",
    system_message=(
        "You are a Retriever. "
        "Your task is to use the 'elasticsearch_hybrid_search' tool to retrieve "
        "candidate profiles from Elasticsearch."
    ),
    llm_config=llm_config,
    functions=[elasticsearch_hybrid_search],
)

researcher = ConversableAgent(
    name="Researcher",
    system_message=(
        "You are a Researcher. "
        "Your role is to use the 'search_in_internet' tool to find individual "
        "job offers related to the candidates profiles. Each job offer must include "
        "a direct link to a specific position, not just a category or group of offers. "
        "Ensure that all job offers are relevant and accurate."
    ),
    llm_config=llm_config,
    functions=[search_in_internet],
)

matcher = ConversableAgent(
    name="Matcher",
    system_message=(
        "Your role is to match job offers with suitable candidates. "
        "The matches must be accurate and beneficial for both parties. "
        "Only match candidates with job offers that fit their qualifications."
    ),
    llm_config=llm_config,
)

critic = ConversableAgent(
    name="Critic",
    system_message=(
        "You are the Critic. "
        "Your task is to verify the accuracy of job-candidate matches. "
        "If the matches are correct, inform the Admin and include the word "
        "'TERMINATE' to end the process."
    ),
    llm_config=llm_config,
)
```

### Running the GroupChat

```python
user = ConversableAgent(name="user", human_input_mode="NEVER")

pattern = AutoPattern(
    initial_agent=admin,
    agents=[admin, retriever, researcher, matcher, critic],
    user_agent=user,
    group_manager_args={"llm_config": llm_config},
)

result, context, last_agent = initiate_group_chat(
    pattern=pattern,
    messages=(
        "Compare the candidate profiles retrieved by the Retriever with the job offers "
        "found by the Researcher on the internet. "
        "Both candidate profiles and job offers are related to the software industry. "
        "Ensure that each match is accurate and beneficial for both parties. "
        "Each candidate should be matched with a single job offer. "
        "Include the job offer link provided by the Researcher."
    ),
    max_rounds=20,
)
```
